```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = df = pd.read_csv('/content/drive/MyDrive/CSVs/Power BI/datos_visualizacion_s3.csv')
df['Fecha']=pd.to_datetime(df['Fecha'])
#display(df.head())

#Gráfico Boxplot
sns.boxplot(x=df['Categoría'],y=df['Precio Unitario'],patch_artist=True,notch=True)
plt.title("\nPrecio Unitario por Categoría")
sns.set_style("darkgrid")
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = df = pd.read_csv('/content/drive/MyDrive/CSVs/Power BI/datos_visualizacion_s3.csv')
df['Fecha']=pd.to_datetime(df['Fecha'])
#display(df.head())
mapa_calor = df.select_dtypes(include=['float64','int64'])
plt.title("\nMatriz de Correlación")
sns.heatmap(mapa_calor.corr(),annot=True,cmap='YlGnBu',fmt='.2f')
plt.show()
```
