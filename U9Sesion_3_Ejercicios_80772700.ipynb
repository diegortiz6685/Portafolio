EJERCICIO 2: Gráfico de Barras Apiladas para Comparar el Gasto Mensual por Región

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://drive.google.com/uc?export=download&id=1VbrQjDF2Br0tjHHiruUnp-0qmnQwon1I')

#3. Agrupar los datos por Región, calculando el Gasto Mensual promedio
prom_region = df.groupby('Region')['Gasto Mensual'].mean()
print(prom_region.round(2))
df_new = pd.DataFrame(prom_region)
transpuesto = df_new.T

#Gráfico de barras apiladas con Seaborn y Matplotlib
print("\nGráfico de Barras Apiladas\n")
transpuesto.plot(kind='bar',stacked=True)
plt.xticks(rotation=0)
plt.title('Gasto Mensual por Región')
plt.show()

#Grafico barras con Seaborn
sns.barplot(df_new, x='Region', y='Gasto Mensual', hue = 'Region')
plt.title('Gasto Mensual por Región')
#plt.xlabel('Región')
#plt.ylabel('Gasto Mensual')
plt.xticks(rotation=0)
print("\nGráfico de Barras Seaborn\n")
plt.show()

#Grafico de barras apiladas con 3 variables
gasto_region = df.groupby(['Region','Edad'])['Gasto Mensual'].mean().reset_index()
pvt = gasto_region.pivot(index='Edad', columns='Region', values='Gasto Mensual')
pvt.plot(kind='bar', alpha = 1, stacked=True)
#sns.barplot(data=gasto_region, x='Region', y='Gasto Mensual', hue='Edad')
plt.title('Gasto Mensual por Región')
plt.xticks(rotation=0,fontsize=8)
plt.ylabel('Gasto Mensual')
print("\nGráfico de Barras Apiladas con 3 variables\n")
plt.show()
```

EJERCICIO 3: Histograma Doble para Visualizar la Distribución del Ingreso y Gasto Mensual

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://drive.google.com/uc?export=download&id=1VbrQjDF2Br0tjHHiruUnp-0qmnQwon1I')

#2. Asegurar que las columnas están definidas con valores numéricos
df.info()

#3. Generar histograma doble
plt.hist(df['Ingreso Mensual'], bins=10, alpha = 0.5, label = 'Ingreso Mensual')
plt.hist(df['Gasto Mensual'], bins=10, alpha = 0.5, label = 'Gasto Mensual')
plt.legend()
print("\nHistograma Doble\n")
plt.show()
```

EJERCICIO 4: Boxplot y Swarmplot para Analizar la Relación entre Puntuación de Crédito e Ingreso Mensual

EJERCICIO 5: KDE Plot para Comparar la Distribución del Ingreso Mensual por Grupos de Edad

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://drive.google.com/uc?export=download&id=1VbrQjDF2Br0tjHHiruUnp-0qmnQwon1I')

#2. Crear nueva columna con rangos de edad
def rango_edad(n):
    if 18 <= n <= 30:
        return "Joven"
    elif 31 <= n <= 50:
        return "Adulto"
    elif 51 <= n <= 65:
        return "Adulto Mayor"
    else:
        return "Fuera de rango"
df['Rangos de Edad'] = df['Edad'].apply(rango_edad)
display(df.head(10))

#3. Crear gráfico KDE Plot
sns.kdeplot(data=df, x='Ingreso Mensual', hue='Rangos de Edad', fill=True)
print("\nGráfico de Densidad\n")
```
