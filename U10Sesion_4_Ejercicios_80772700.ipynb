```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt

df = df = pd.read_csv('/content/drive/MyDrive/CSVs/Power BI/datos_visualizacion_s3.csv')
#display(df.head(10))
#print("\nINFORMACIÓN DEL DATASET")
df['Fecha']=pd.to_datetime(df['Fecha'])
df.info()

#Gráfico de Ventas por Categoría
#plt.figure(figsize=(8,6))
ventas_cat = df.groupby('Categoría')['Total Ventas'].sum()
ventas_cat = ventas_cat.sort_values(ascending=True)
ventas_cat.plot(kind='barh',color='pink')
plt.title('Ventas por Categoría')
plt.xlabel('Total Ventas')
plt.ylabel('Categoría')
print("\nGRÁFICO DE VENTAS POR CATEGORÍA")
plt.show()

#Gráfico de Ventas en el tiempo
plt.figure(figsize=(12,6))
df['Mes_Año'] = df['Fecha'].dt.strftime('%Y-%m')
ventas_tiempo = df.groupby('Mes_Año')['Total Ventas'].sum()
#ventas_tiempo = df.groupby('Fecha')['Total Ventas'].sum()
ventas_tiempo.plot(kind='line',color='green',marker="o",ms=8)
plt.title('Ventas en el tiempo')
plt.xlabel('Fecha')
plt.ylabel('Total Ventas')
plt.grid(color='gray',linestyle='--',linewidth=0.8)
print("\n GRÁFICO DE VENTAS EN EL TIEMPO\n")
plt.show()
```
