EJERCICIO 1: Visualización de Ventas por Categoría con un Gráfico de Barras

```python
import pandas as pd
import matplotlib.pyplot as plt

#1. Crear DataFrame del CSV
df = pd.read_csv('https://drive.google.com/uc?export=download&id=1JS7jkpbWMaOlpsh_7Q6P8nue1U9aiM-C')

#2. Agrupar datos por Categoría y sumar cantidades vendidas
suma_cat = df.groupby('Categoría')['Cantidad Vendida'].sum()

#3. Generar gráfico de barras
suma_cat.plot(kind='bar')
plt.title('Cantidad Vendida por Categoría')
plt.xlabel('Categoría')
plt.xticks(rotation=30)
plt.ylabel('Cantidad Vendida')

#4. Mostrar Resultados
print("\nLas Cantidades vendidas por Categoría son:\n")
print(suma_cat,"\n")
plt.show()
```

EJERCICIO 2: Análisis de Ventas a lo Largo del Tiempo con un Gráfico de Líneas

```python
import pandas as pd
import matplotlib.pyplot as plt

#1. Crear DataFrame del CSV
df = pd.read_csv('https://drive.google.com/uc?export=download&id=1JS7jkpbWMaOlpsh_7Q6P8nue1U9aiM-C')

#2. Convertir Columna Fecha a Formato DateTime
df['Fecha'] = pd.to_datetime(df['Fecha'])

#3. Agrupar datos por Fecha y sumar cantidades vendidas
suma_fecha = df.groupby('Fecha')['Cantidad Vendida'].sum()

#4. Generar gráfico de líneas
suma_fecha.plot(kind='line',color='red')
plt.title('Cantidad Vendida por Categoría')
plt.xlabel('Categoría')
plt.xticks(rotation=30)
plt.ylabel('Cantidad Vendida')
plt.grid(True)

#4. Mostrar Resultados
print("\nLas Cantidades vendidas por Categoría son:\n")
print(suma_fecha,"\n")
plt.show()
```

EJERCICIO 3: Visualización de la Distribución de Precios Unitarios con un Histograma

```python
import pandas as pd
import matplotlib.pyplot as plt

#1. Crear DataFrame del CSV
df = pd.read_csv('https://drive.google.com/uc?export=download&id=1JS7jkpbWMaOlpsh_7Q6P8nue1U9aiM-C')

#2. Analizar Distribución de Precio Unitario por rangos de precio
rangos = pd.cut(df['Precio Unitario'], bins=10)
distribucion = rangos.value_counts().sort_index()
print("\nLa Distribución de Precio Unitario es:\n")
print(distribucion.round(2),"\n")

#3. Generar Histograma

plt.hist(df['Precio Unitario'], bins=10, edgecolor='black')
plt.title('Distribución de Precios Unitarios')
plt.xlabel('Precio Unitario')
plt.ylabel('Frecuencia')

plt.show()
```

EJERCICIO 4: Comparación de Ventas Totales por Producto con un Gráfico de Barras Horizontal

```python
import pandas as pd
import matplotlib.pyplot as plt

#1. Crear DataFrame del CSV
df = pd.read_csv('https://drive.google.com/uc?export=download&id=1JS7jkpbWMaOlpsh_7Q6P8nue1U9aiM-C')

#2. Agrupar datos por producto y calcular ventas totales
ventas_totales = df.groupby('Producto')['Cantidad Vendida'].sum()

#3. Generar gráfico de barras horiziontales
ventas_totales.plot(kind='barh')
plt.title('Ventas Totales por Producto')
plt.xlabel('Ventas Totales')
plt.ylabel('Producto')

#4. Mostrar Resultados
print("\nLas ventas totales por producto son:\n")
print(ventas_totales,"\n")
plt.show()
```
